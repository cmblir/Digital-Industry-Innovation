```python
def make_ev_dict(df) -> pd.DataFrame:
    ev_dict = {}
    for key, value in zip(df["표준컬럼 영문명"], df["표준 데이터타입"]):
        ev_dict[key] = value
    columns = [x.lower() for x in df["표준컬럼 영문명"]]
    return ev_dict, pd.DataFrame(columns=columns)
```

```python
def append_data(base, df):
    for idx in tqdm(range(len(base))):
        symbol = base["lstng_cd"][idx]
        append_df = df[df["lstng_cd"] == symbol]
        if not append_df.empty:
            base.loc[idx, list(append_df.columns)] = append_df.values[0]
    return base
```
