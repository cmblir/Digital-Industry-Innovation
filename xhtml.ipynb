```python
from bs4 import BeautifulSoup

with open("sample.xhtml", "r", encoding="utf-8") as file:
    xhtml_content = file.read()

soup = BeautifulSoup(xhtml_content, features="xml")

con_fraction_elements = soup.find_all("ix:nonFraction")
```

```python
for element in con_fraction_elements:
    name = element.get("name")
    print(f"태그 이름: {element.name}, name 속성: {name.replace('core:', '')}, 값 : {element.text}")
```
